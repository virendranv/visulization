# Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
x = [2, 4, 6]
y = [1, 3, 5]
plt.plot(x, y)
plt.show()
```

```python
year = [1960, 1970, 1980, 1990, 2000, 2010]
pop_us = [44.91, 58.09, 78.07, 107.7, 138.5, 170.6]
pop_india = [449.48, 553.57, 696.783, 870.133, 1000.4, 1309.1]
plt.plot(year, pop_us, color='g',label='US')
plt.plot(year, pop_india, color='orange',label='india')
plt.xlabel('Countries')
plt.ylabel('Population in million')
plt.title(' India _ US Population till 2010')
plt.legend()
plt.show()
```

# Bar chart

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Make a fake dataset:
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
 
# Create bars
plt.bar(bars, height)
 

# Show graphic
plt.show()
```

# pie chart

```python
import matplotlib.pyplot as plt
 
import matplotlib.pyplot as plt
 
# Data to plot
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

gaussian_numbers = np.random.randn(1000)
plt.hist(gaussian_numbers)
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
```

Graphs:
               
               Description
               
distplot   -   histogram
barplot    -   estimate of central tendency for a numeric variable
violinplot -   similar to boxplot,also shows the probability density of the data

lmplot     -   Regression plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tip=sns.load_dataset('tips')
```

```python
tip.head()
```

# Distribution Plot

Distribution plot will gives univariate

In histogram on top of that line is known as Kernal Density Function(KDE)


```python
sns.distplot(tip['total_bill'],kde = True)
```

# Joint Plot

Joint plot is bivariate .we can validate how the two variables distributed


```python
sns.jointplot(x='total_bill',y='tip',data = tip)
```

# Pair Plot

Pair plot will plot a diagram all against the all possible combination of all numerical values. This  is most important plots

We will work on this plot through  out our ML


```python
sns.pairplot(data=tip)
```

# Rug plot

Distplot will shows the bins .Rugplot will shows the itâs points of the line.  Rest  all are same


```python
sns.rugplot(tip['total_bill'])
```

# Bar plot

```python
sns.barplot(x='sex',y='total_bill',data=tip)
```

# Count plot

```python
sns.countplot(x='sex',data=tip)
```

# Box plot

Box plot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution,except for points that are determined to be âoutliersâ 


```python
sns.boxplot(data=tip)
```

# Violinplot

A violinplot plays a similar role as aboxplot and whiskerplot it shows the distribution of  dataset


```python
sns.violinplot(data=tip)
```

# strip plot

```python
sns.stripplot(x='day',y='total_bill',data=tip)
```

# Swarm plot

```python
sns.swarmplot(data=tip,x='sex',y='total_bill')
```

# Factorplot

Factor plot is the most general form of a categorical and numerical plot.it can take in a kind parameter to adjust the plot type


```python
sns.factorplot(data=tip,x='day',y='total_bill')
```
